```
img = cv2.imread('/content/dogs.jpg')
```

```
print(img.shape)
```

```
plt.imshow(img)
plt.show()
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

- Flip Upside Down and display it in the notebook

```
img_flip = cv2.flip(img_rgb, 0)
plt.imshow(img_flip)
```

- Draw an empty RED rectangle around dogs face and display it

```
img_rec2 = cv2.rectangle(img_rgb, (420,420), (650,180), (255,0,0), 3)
plt.imshow(img_rec2)
```

- Draw a BLUE TRIANGLE in the middle of the image, and display it

```
img_tri = cv2.line(img_rgb, (420,420), (560,180), (0,0,255), 3)
img_tri = cv2.line(img_rgb, (560,180), (650,420), (0,0,255), 3)
img_tri = cv2.line(img_rgb, (650,420), (420,420), (0,0,255), 3)
plt.imshow(img_tri)
```

- fill in the triangle
  - `cv2.fillPoly()`

```
pts = np.array([[420, 420], [560, 180], [650, 420]], np.int32)
print(pts.shape)

img = cv2.fillPoly(img_rgb, [pts], (0, 0, 255))
```

```
plt.imshow(img)
plt.show()
```
